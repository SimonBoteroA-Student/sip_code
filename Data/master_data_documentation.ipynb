## Resumen de Contenidos por Carpeta

### `Propia/` - Datos Propios y Consolidados

**`Contraloria Data Merger/`**: Datos consolidados de Contraloría
- `boletines.csv`: Responsables fiscales extraídos de boletines PDF

**`Datos_Abiertos/`**: Datos públicos de contratación SECOP (gobierno abierto)
- Información completa del ciclo de contratación pública
- Procesos, contratos, ofertas, ejecución, sanciones

**`PACO/`**: Proyecto Anticorrupción de la Contraloría
- Consolidación de múltiples fuentes (SECOP, SIRI, Procuraduría, SIC, FGN)
- Información de obras inconclusas, sanciones disciplinarias y penales

**`Monitor/`**: Base de datos de hechos y actores para seguimiento
- Información estructurada para análisis de corrupción (2016-2022)

### `PredictingMisconduct/` - Análisis Predictivo

Modelo de predicción de conducta indebida con múltiples características:
- Demográficas, historia (HK), recursos naturales, actos ilegales, sector financiero, sector público, conflictos, crimen, política, actividad económica

### `Vigia/` - Sistema de Vigilancia

Datos crudos y procesados de SECOP II con modelos predictivos entrenados.

## 3. `Contraloria Data Merger/extract_boletines.py`

Script en Python que **genera** `boletines.csv` a partir de boletines PDF de la Contraloría. No es un archivo tabular.

**Localización**: `Contraloria Data Merger/`

**Funcionalidad**:
- Lee archivos PDF de boletines de la Contraloría.
- Detecta tablas y extrae sus celdas.
- Limpia y normaliza las columnas de salida (unificación de tipo y número de documento, normalización de espacios, etc.).
- Define el **orden de columnas de salida** (las de `boletines.csv`).
- Genera el archivo `boletines.csv` con la estructura especificada.

Para revisar la lógica de extracción detallada se debe leer directamente el código del script.
